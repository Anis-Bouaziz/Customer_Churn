# <font color = red>1 - Data Visualization</font>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.basemap import Basemap
```

```python
#df = pd.read_csv('Churn_vis.csv',index_col=0)
df = pd.read_excel('Telco_customer_churn.csv.xlsx')
sns.set()
```

### <font color='green'> Plotting Spatial Data on California Map </font>

```python
fig, ax = plt.subplots(figsize = (12,10))
BBox = ((df.Longitude.min(),   df.Longitude.max(),      
         df.Latitude.min(), df.Latitude.max()))
ruh_m = plt.imread('map(2).png')
ax.scatter(df.Longitude, df.Latitude, zorder=0.1, alpha= 0.5, s=10,facecolors='none', edgecolors='b')
ax.set_title('Plotting Spatial Data on California Map')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(ruh_m, zorder=0, extent = BBox, aspect= 'equal')
```

### <font color='red'> Note </font>
#### Customers who have left are distributed randomly except in a few cities (but that's simply because there are more customers in those cities).


### <font color='green'> Churn Count </font>

```python
explode = (0,0.05)
plt.figure(figsize = (8,8))
plt.pie(df['Churn Value'].value_counts(),explode = explode,shadow = True,autopct='%1.2f%%',colors = ['pink','teal'])
plt.legend(['Non Churn Customer','Churn Customer'])
plt.title('Customer Churn')
plt.show()
```

#### Customers who churned represent 26.58% of the data

### <font color='green'> Churn probability by Payment method and Gender

```python
fig, ax = plt.subplots(figsize=(12,12))
sns.barplot(ax=ax,x="Payment Method", y="Churn Value",hue="Gender", data=df)
plt.ylabel('Churn probability')
plt.title('Churn probability by Payment method and Gender',fontdict = {'fontsize' : 20});
```

#### we can see that , no matter their gender,customers who use Electronic check as a payment method have a higher probability of churning 

### <font color='green'> Churn count based on gender , age , and total charges paid

```python
fig,axes=plt.subplots(1,2,sharey=True,figsize=(20, 8))
df['Total Charges']= pd.to_numeric(df['Total Charges'], errors='coerce')
sns.violinplot(ax=axes[0],data=df,x='Churn Value',y='Total Charges',hue='Gender',split=True)
sns.violinplot(ax=axes[1],data=df,x='Churn Value',y='Total Charges',hue='Senior Citizen',split=True)
axes[0].set_xlabel('Churn')
axes[1].set_xlabel('Churn')
axes[0].set_xticklabels(['no','yes'])
axes[1].set_xticklabels(['no','yes'])
axes[0].set_title('Churn count by Gender and Total charges')
axes[1].set_title('Churn count by Gender and age');
```

#### customers with lower total charges tend to churn more 
#### gender has no significance on churn probability 
#### younger citizens tend to churn more than senior citizens

### <font color='green'> Churn Probability based on Contract type

```python
g1=sns.catplot(x="Contract", y="Churn Value", data=df,kind="bar")
g1.set_ylabels("Churn Probability")
df.groupby(['Contract']).agg({"Churn Value" : (np.sum)})
plt.title('Churn Probability based on Contract type')
```

#### We can see that customers with a month to month contract have a higher chance of churning and those with a long term contract don't churn alot .

### <font color='green'> Churn Probability based on Internet and phone service

```python
fig,axes=plt.subplots(1,2,sharey=True,figsize=(20, 8))
g2=sns.barplot(ax=axes[0],x="Internet Service", y="Churn Value", data=df)
axes[0].set_ylabel("Churn Probability")
g3=sns.barplot(ax=axes[1],x="Phone Service", y="Churn Value", data=df)
axes[1].set_ylabel("Churn Probability");
axes[1].set_title('Churn Probability based on phone service')
axes[0].set_title('Churn Probability based on Internet service');
```

#### Customers who paid for a fiber optic service surprisingly tend to churn more , and those with no internet service at all churn the least .
#### having a phone service has no important impact on churning 

### <font color='green'> Churn Probability based on Different charges

```python
sns.pairplot(df,vars = ['Tenure Months','Monthly Charges','Total Charges'], hue="Churn Value")
```

#### Pair plots allow us to see both distribution of single variables (Tenure ,Monthly Charges,Total Charges) and relationships between two variables.


### <font color='green'> Correlation Matrix

```python
f, ax=plt.subplots(figsize=(11,11))
sns.heatmap(df.corr(),annot=True,cmap="coolwarm")
```

```python
data=pd.get_dummies(df[['Gender','Senior Citizen','Partner','Dependents','Phone Service','Multiple Lines','Internet Service','Online Security','Online Backup','Device Protection','Tech Support','Streaming TV','Streaming Movies','Contract','Paperless Billing','Payment Method','Churn Value']])
plt.figure(figsize=(10,15))
sns.heatmap(data.corr()[['Churn Value']].sort_values(by='Churn Value',ascending=False),cmap="RdBu_r",center=0.0, annot=True);
```

#### finally we wanted to explore the correlation between different charges (correlation matrix ) and the relationship between the different categorical data and our target variable (correlation heatmap) 
#### we can see there's a strong correlation between the churn score and churning and that's because that score is a statistical machine learning prediction value that estimates the state of churn for a given user at any given time, evaluating user demographic info, browsing behavior, and historical purchase data among other signals , so it has a strong influence on our desired output.
